```python
h_df = aapl_df.copy() 
d_df = aapl_historical.copy()
h_df.drop(columns = ['Dividends','Stock Splits'],axis=1,inplace=True)  
d_df.drop(columns = ['Stock Splits'],axis=1,inplace=True)  
```

## Average Hourly Price and Movement 

```python
# Calculating the pct change in high/low price by hour
h_df['Hourly Movement'] = h_df[['Low', 'High']].pct_change(axis=1)['High'] 

# Calulting the Average Hourly Price using Avg Hourly Movement to Calculate Estimated Market Volume by Hour
h_df['Trade Value'] = ((h_df['Low']*h_df['Hourly Movement']) * h_df['Volume'] ) 

h_df['Date'] = h_df.index.date 

agg_df = h_df[['Date', 'Hourly Movement','Trade Value']].copy().reset_index()  

# Average pct change in high low price for each day / total monetary volume by day 
avg_df = agg_df.groupby(['Date']).agg({'Hourly Movement':'mean','Trade Value':'sum'}).reset_index()   

avg_df = avg_df.rename({'Hourly Movement': "Avg Hourly High/Low Pct Change",'Trade Value':'Market Volume'},axis=1)  

avg_df = avg_df.set_index('Date') 

print(avg_df.shape)
avg_df.head()
```

## Data Freq = Day to be merged with Engineered Features

```python
d_df['Day'] = d_df.index.date
```

```python

temp_df = d_df.reset_index()
aapl_df = temp_df.drop(columns=['Date'],axis=1,inplace=True) 

aapl_day = temp_df.rename({'Day':'Date'},axis=1)    

aapl_day = aapl_day.set_index('Date')  

print(aapl_day.shape) 
aapl_day.head()
```

```python
merged_df = pd.concat([aapl_day,avg_df],axis=1)  

df_new = merged_df.loc[:,merged_df.columns!='Close'] 

df_new['Close'] = merged_df['Close']   

df_new.head()
```
